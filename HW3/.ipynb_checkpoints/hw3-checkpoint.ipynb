```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler 
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
from sklearn.neighbors import KNeighborsClassifier
```

```python
train = pd.read_csv(r"C:\Users\navee\Downloads\ML Materials\HW3\optdigits.tra", header =None)
test = pd.read_csv(r"C:\Users\navee\Downloads\ML Materials\HW3\optdigits.tes", header =None)

train_x = train.iloc[:,0:64].values
train_y = train.iloc[:,[64]].values
test_x = test.iloc[:,0:64].values
test_y = test.iloc[:,[64]].values
print(train_x.shape)
print(train_y.shape)
train.describe()[1:3]
```

```python
np.unique(train_y, return_counts=True)
np.unique(test_y, return_counts=True)
```

# Perform PCA for K = 2, 3, 8 & 20

```python


#Scale the data first
sc = StandardScaler()
train_x = sc.fit_transform(train_x)
test_x = sc.fit_transform(test_x)

#PCA with k = 2,3,8 & 20
pca2 = PCA(n_components = 2)
pca3 = PCA(n_components = 3)
pca8 = PCA(n_components = 8)
pca20 = PCA(n_components = 20)

#calculate principal components for traininig data
train_pca2 = pca2.fit_transform(train_x)
train_pca3 = pca3.fit_transform(train_x)
train_pca8 = pca8.fit_transform(train_x)
train_pca20 = pca20.fit_transform(train_x)

#calculate principal components for test data
test_pca2 = pca2.fit_transform(test_x) 
test_pca3 = pca3.fit_transform(test_x)
test_pca8 = pca8.fit_transform(test_x)
test_pca20 = pca20.fit_transform(test_x)
```

# 1- Nearest Neighbor Classification for PCA

```python
def knn_classifier(train_x,train_y,test_x, k):
    
    #define KNN for k =1
    classifier = KNeighborsClassifier(n_neighbors=1)
    #fit the classifier on training data with pca2
    classifier.fit(train_x, train_y)
    #predict on training data
    train_pred = classifier.predict(train_x)
    #predict on test data with pca2
    pred = classifier.predict(test_x) 
    #create dataframe with real target and predicted target
    pred_pca = pd.concat([test[64], pd.DataFrame(pred, columns=["pred_pca" + str(k)])], axis = 1)
    train_pred_pca = pd.concat([train[64],pd.DataFrame(train_pred, columns=["pred_pca" + str(k)])], axis =1)

    #record the error for each class 
    test_error_pca = pd.DataFrame()
    train_error_pca = pd.DataFrame()
    for c in pred_pca[64].unique():
        classtest = pred_pca[pred_pca[64]== c]
        classtrain = train_pred_pca[train_pred_pca[64]==c]
        test_error_pca[str(c)] = [np.mean(classtest[64] != classtest["pred_pca" + str(k)])]
        train_error_pca[str(c)] = [np.mean(classtrain[64] != classtrain["pred_pca" + str(k)])]

    return (test_error_pca, train_error_pca)
```

```python
train_error = pd.DataFrame()
test_error = pd.DataFrame()

(test_2,train_2) = knn_classifier(train_pca2,train_y,test_pca2, 2)
(test_3,train_3) = knn_classifier(train_pca3,train_y,test_pca3, 3)
(test_8,train_8) = knn_classifier(train_pca8,train_y,test_pca8, 8)
(test_20,train_20) = knn_classifier(train_pca20,train_y,test_pca20, 20)

train_error = pd.concat([train_2,train_3, train_8, train_20], axis = 0)
test_error = pd.concat([test_2, test_3, test_8, test_20],axis = 0)

train_error.to_csv(r"C:\Users\navee\Downloads\ML Materials\HW3\pca_knn_train_error.csv", index= False)
test_error.to_csv(r"C:\Users\navee\Downloads\ML Materials\HW3\pca_knn_test_error.csv", index= False)
```

# LDA for k = 2, 3, 8 & 20

```python
#LDA with k = 2,3,8 & 20
lda2 = LDA(n_components=2)
lda3 = LDA(n_components=3)
lda8 = LDA(n_components=8)
lda20 = LDA(n_components=20)

#calclulate linear discriminants for training data
train_lda2 = lda2.fit_transform(train_x, train_y)
train_lda3 = lda3.fit_transform(train_x, train_y)
train_lda8 = lda8.fit_transform(train_x, train_y)
train_lda20 = lda20.fit_transform(train_x, train_y)

#calculate linear discriminant for test data
test_lda2 = lda2.transform(test_x)
test_lda3 = lda3.transform(test_x)
test_lda8 = lda8.transform(test_x)
test_lda20 = lda20.transform(test_x)
```

# 1- Nearest Neighbor Classification for LDA

```python
def knn_classifier_lda(train_x,train_y,test_x, k):
    
    #define KNN for k =1
    classifier = KNeighborsClassifier(n_neighbors=1)
    #fit the classifier on training data with lda2
    classifier.fit(train_x, train_y)
    #predict on training data
    train_pred = classifier.predict(train_x)
    #predict on test data with lda2
    pred = classifier.predict(test_x) 
    #create dataframe with real target and predicted target
    pred_lda = pd.concat([test[64], pd.DataFrame(pred, columns=["pred_lda" + str(k)])], axis = 1)
    train_pred_lda = pd.concat([train[64],pd.DataFrame(train_pred, columns=["pred_lda" + str(k)])], axis =1)

    #record the error for each class 
    test_error_lda = pd.DataFrame()
    train_error_lda = pd.DataFrame()
    for c in pred_lda[64].unique():
        classtest = pred_lda[pred_lda[64]== c]
        classtrain = train_pred_lda[train_pred_lda[64]==c]
        test_error_lda[str(c)] = [np.mean(classtest[64] != classtest["pred_lda" + str(k)])]
        train_error_lda[str(c)] = [np.mean(classtrain[64] != classtrain["pred_lda" + str(k)])]

    return (test_error_lda, train_error_lda)

train_error = pd.DataFrame()
test_error = pd.DataFrame()

(test_2,train_2) = knn_classifier_lda(train_lda2,train_y,test_lda2, 2)
(test_3,train_3) = knn_classifier_lda(train_lda3,train_y,test_lda3, 3)
(test_8,train_8) = knn_classifier_lda(train_lda8,train_y,test_lda8, 8)
(test_20,train_20) = knn_classifier_lda(train_lda20,train_y,test_lda20, 20)

train_error = pd.concat([train_2,train_3, train_8, train_20], axis = 0)
test_error = pd.concat([test_2, test_3, test_8, test_20],axis = 0)

train_error.to_csv(r"C:\Users\navee\Downloads\ML Materials\HW3\lda_knn_train_error.csv", index= False)
test_error.to_csv(r"C:\Users\navee\Downloads\ML Materials\HW3\lda_knn_test_error.csv", index= False)
```
