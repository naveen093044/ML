```python
import pandas as pd
import numpy as np
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier 
from sklearn.tree import DecisionTreeRegressor
from sklearn import metrics
import matplotlib.pyplot as plt 
import xgboost as xgb
import lightgbm as lgb
from sklearn.neighbors import KNeighborsClassifier
```

# Import data

```python
train = pd.read_csv(r"C:\Users\navee\Downloads\ML Materials\HW4\optdigits.tra", header =None)
test = pd.read_csv(r"C:\Users\navee\Downloads\ML Materials\HW4\optdigits.tes", header =None)
```

# Create X_25 and X_100 datasets by choosing random data points

```python
#creating the dataset X_25 containing 25 randomly selected data points from each class and X_100 contaning 100 from each class

temp = train[train[64]==0]
X_25 = temp.iloc[np.random.randint(0,temp.shape[0], size = 25),:]
X_100 = temp.iloc[np.random.randint(0,temp.shape[0], size = 100),:]
for c in range(1,10):
    temp = train[train[64]==c]
    x25 = temp.iloc[np.random.randint(0,temp.shape[0], size = 25),:]
    x100 = temp.iloc[np.random.randint(0,temp.shape[0], size = 100),:]
    X_25 = X_25.append(x25, ignore_index=True)
    X_100 = X_100.append(x100, ignore_index=True)
```

# Create DecisiontreeClassifier for depth = [2, 3, 5, 10]

```python
depth = [2, 3, 5, 10]
dt_performance = {}

for i in depth:
    #initialize the classifier with max depth i 
    clf = DecisionTreeClassifier(max_depth=i)
    #train on the X_25 dataset
    clf.fit(X_25.iloc[:,0:64],X_25[[64]])
    #predict for training data and record train error for depth i
    x_25_train = clf.predict(X_25.iloc[:,0:64])
    dt_performance["train_err_25_" + str(i)] = 1 - metrics.accuracy_score(x_25_train, X_25[[64]])
    #predict for test data and record test error for depth i
    x_25_test = clf.predict(test.iloc[:,0:64])
    dt_performance["test_err_25_" + str(i)] = 1 - metrics.accuracy_score(x_25_test, test[[64]])
   
    #train on X_100 dataset
    clf.fit(X_100.iloc[:,0:64],X_100[[64]])
    #predict for training data and record train error for depth i
    x_100_train = clf.predict(X_100.iloc[:,0:64])
    dt_performance["train_err_100_" + str(i)] = 1 - metrics.accuracy_score(x_100_train, X_100[[64]])
    #predict for test data and record test error for depth i
    x_100_test = clf.predict(test.iloc[:,0:64])
    dt_performance["test_err_100_" + str(i)] = 1 - metrics.accuracy_score(x_100_test, test[[64]])
    print("tree for depth = ",i)
    tree.plot_tree(clf)
    plt.show()

    
print(dt_performance)
```

# KNN for k = [1,3,5,9]

```python
k = [1,3,5,9]
knn_train_err_25 = []
knn_test_err_25 = []
knn_train_err_100 = []
knn_test_err_100 = []
for i in k:
    #initialize the classifier for k = i 
    clf = KNeighborsClassifier(n_neighbors = i)
    #train on the X_25 dataset
    clf.fit(X_25.iloc[:,0:64],X_25[[64]])
    #predict for training data and record train error for  k =i 
    x_25_train = clf.predict(X_25.iloc[:,0:64])
    knn_train_err_25.append( 1 - metrics.accuracy_score(x_25_train, X_25[[64]]))
    #predict for test data and record test error for k= i
    x_25_test = clf.predict(test.iloc[:,0:64])
    knn_test_err_25.append( 1 - metrics.accuracy_score(x_25_test, test[[64]]))
   
    #train on X_100 dataset
    clf.fit(X_100.iloc[:,0:64],X_100[[64]])
    #predict for training data and record train error for k=i
    x_100_train = clf.predict(X_100.iloc[:,0:64])
    knn_train_err_100.append( 1 - metrics.accuracy_score(x_100_train, X_100[[64]]))
    #predict for test data and record test error for k=i
    x_100_test = clf.predict(test.iloc[:,0:64])
    knn_test_err_100.append( 1 - metrics.accuracy_score(x_100_test, test[[64]]))
    

```

```python
plt.plot(k, knn_test_err_25)
plt.title('Test errors for different k in knn for X_25 dataset')
plt.xlabel('k')
plt.ylabel('error')
plt.savefig(r"C:\Users\navee\Downloads\ML Materials\HW4\knn_test_err_25.jpg")
```

```python
plt.plot(k, knn_test_err_100)
plt.title('Test errors for different k in knn for X_100 dataset')
plt.xlabel('k')
plt.ylabel('error')
plt.savefig(r"C:\Users\navee\Downloads\ML Materials\HW4\knn_test_err_100.jpg")
```

# Decision tree Regression 

```python
test_error = []
for i in range(48,64): # i will have the value from 48 to 64
    #initialize regressor
    rg = DecisionTreeRegressor(max_depth= 10)
    #train the model on X_100_69 data for first 48 columns and ith column as ground truth
    rg.fit(X_100_69.iloc[:,0:48],X_100_69[[i]])
    pred = rg.predict(test_69.iloc[:,0:48])
    test_error.append(metrics.mean_squared_error(test_69[[i]], pred))
    
    
    
```

```python
plt.plot(range(48,64),test_error)
plt.title('Test error(MSE) for predicting pixels 48 to 64')
plt.xlabel('Pixels(columns)')
plt.ylabel('error')
plt.savefig(r"C:\Users\navee\Downloads\ML Materials\HW4\dt_test_err_100.jpg")
```
